## **1. Import Library**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import time
```

PENJELASAN:
1. import requests


- Fungsi: Untuk mengirimkan permintaan HTTP (seperti GET atau POST) ke server web.

- Contoh: Digunakan untuk mengambil kode HTML dari sebuah halaman web yang akan di-scrape.

2. from bs4 import BeautifulSoup

- Fungsi: BeautifulSoup adalah library untuk parsing (mengurai) kode HTML atau XML. Ini memudahkan ekstraksi informasi spesifik dari halaman web, misalnya mencari semua tabel, tautan, atau teks dengan kelas tertentu.

- Catatan: bs4 adalah nama paketnya, dan BeautifulSoup adalah kelas utama yang akan digunakan.

3. import pandas as pd

- Fungsi: Pandas adalah library untuk manipulasi dan analisis data, terutama dengan struktur DataFrame (mirip tabel/spreadsheet). Ini sangat berguna untuk membersihkan data, menggabungkan tabel, dan menyimpan hasil scraping (misalnya ke file CSV atau Excel).

4. import json

- Fungsi: Untuk memproses data dalam format JSON (JavaScript Object Notation). Banyak situs modern mengirimkan data melalui API dalam format JSON. Library ini digunakan untuk mengubah string JSON menjadi dictionary Python (atau sebaliknya).

5. import time

- Fungsi: Menyediakan fungsi-fungsi yang berkaitan dengan waktu, seperti time.sleep(). Ini penting dalam web scraping untuk membuat jeda antar permintaan agar tidak membebani server atau menghindari pemblokiran IP.

## **2. Menentukan URL dan Variabel Penampung**

```python
base_url = "https://loginmegastore.com/search?query=kulkas&p=true&page="
data_list = []
```

PENJELASAN:

Program ini memulai dengan mengimpor lima library:* requests *untuk mengambil halaman web, *BeautifulSoup* untuk membaca dan mencari data di dalam halaman tersebut,* pandas *untuk menyusun data menjadi bentuk tabel,* json *untuk memproses data berformat JSON, dan *time *untuk mengatur jeda waktu agar tidak mengganggu server. Selanjutnya, program menyiapkan alamat dasar pencarian toko online *loginmegastore.com *untuk kata kunci "kulkas" dengan parameter halaman yang masih kosong, serta menyediakan sebuah list kosong bernama *data_list* yang akan digunakan sebagai tempat menampung semua data produk yang akan diambil nanti.

## **3. Proses Scraping Halaman 1–5**

```python
for page in range(1, 6):
    url = base_url + str(page)
    print(f"Sedang scraping halaman {page}...")

    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
```

PENJELASAN:

1. **Perulangan halaman** – Program akan mengulang proses sebanyak 5 kali, dimulai dari halaman 1 sampai halaman 5 (*for page in range(1, 6)*).

2. **Membuat URL lengkap** – Setiap kali perulangan, nomor halaman digabungkan dengan URL dasar* (base_url + str(page)*). Contohnya:

- Halaman 1 menjadi* ...&page=1*

- Halaman 2 menjadi ...&page=2

- Begitu seterusnya hingga halaman 5.

3. **Memberi tahu proses** – print(f"Sedang scraping halaman {page}...") menampilkan teks ke layar agar kita tahu halaman mana yang sedang diproses.

4. **Mengambil halaman web** – *requests.get(url*) berfungsi seperti kita mengetik URL di browser dan mengirim permintaan ke server website.

5. **membaca isi halaman** – *BeautifulSoup(response.content, "html.parser")* mengubah kode HTML yang diterima menjadi struktur data yang mudah dibaca dan dicari.

6. **Hasil yang terlihat** – Setelah program dijalankan, muncul tulisan:

- "Sedang scraping halaman 1..."

- "Sedang scraping halaman 2..."

- ... hingga halaman 5.





## **4. Mengambil Data Sesuai Soal**

```python
    titles = soup.find_all("p", class_="mantine-focus-auto !leading-[120%] m_b6d8b162 mantine-Text-root")
    prices = soup.find_all("p", class_="mantine-focus-auto !whitespace-nowrap m_b6d8b162 mantine-Text-root")
    images = soup.find_all("img", class_="m_9e117634 mantine-Image-root")
```

PENJELASAN:

Kode ini mengambil semua judul produk, harga produk, dan gambar produk dari halaman web berdasarkan kelas CSS spesifik yang dimiliki masing-masing elemen.

## **5. Menggabungkan Data**

```python
    for i in range(len(titles)):
        try:
            title = titles[i].text.strip()
            price = prices[i].text.strip()
            image = images[i]['src']

            data_list.append({
                "title": title,
                "price": price,
                "image": image
            })
        except:
            continue

    time.sleep(2)
```

PENJELASAN:

Kode ini mengambil setiap judul, harga, dan gambar dari hasil pencarian, menggabungkannya menjadi satu record produk, menyimpannya ke dalam *data_list, *lalu berhenti 2 detik sebelum melanjutkan ke halaman berikutnya untuk menghindari pemblokiran server.



## **6. Simpan ke JSON (List_Elektronik.json)**

```python
with open("List_Elektronik.json", "w", encoding="utf-8") as f:
    json.dump(data_list, f, indent=4, ensure_ascii=False)

print("JSON berhasil dibuat")
```

PENJELASAN:

Kode ini menyimpan seluruh data produk yang sudah di-scrape ke dalam file *List_Elektronik.json* dengan format yang rapi dan mendukung karakter Indonesia, lalu menampilkan pesan bahwa file JSON berhasil dibuat.

## **7. Proses ETL (JSON → CSV)**

```python
with open("List_Elektronik.json", "r", encoding="utf-8") as f:
    data = json.load(f)
df = pd.DataFrame(data)
df = df.dropna()
df.to_csv("List_Elektronik.csv", index=False)

print("CSV berhasil dibuat")
```

PENJELASAN:

Kode ini melakukan **proses ETL (Extract, Transform, Load) yaitu **mengambil data dari file JSON, mengubahnya menjadi format tabel, membersihkannya, lalu menyimpannya sebagai file CSV.

 **ETL = Extract → Transform → Loadt**


- **Extract** : Mengambil data dari JSON

- **Transform** : Mengubah ke DataFrame & membersihkan data

- **Load**: Menyimpan ke CSV
